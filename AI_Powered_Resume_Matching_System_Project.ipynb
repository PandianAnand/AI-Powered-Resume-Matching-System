```python
def search_resumes(job_description, top_k):
    query_vector = model.encode(job_description)

    results = client.search(
        collection_name="resumes",
        data=[query_vector],
        limit=top_k
    )

    return results
```

```python
def get_ordinal(n):
    # Special cases for 11, 12, and 13 which end in 'th'
    if 11 <= n % 100 <= 13:
        suffix = 'th'
    else:
        # Dictionary approach for the last digit
        suffixes = {1: 'st', 2: 'nd', 3: 'rd'}
        suffix = suffixes.get(n % 10, 'th')

    return f"{n}{suffix}"
```

```python
import spacy
from sklearn.metrics.pairwise import cosine_similarity

nlp = spacy.load("en_core_web_sm")

def extract_phrases(text):
    doc = nlp(text)

    phrases = set()

    # noun chunks = best “skill-like” candidates
    for chunk in doc.noun_chunks:
        phrase = chunk.text.lower().strip()
        if 2 <= len(phrase.split()) <= 5:
            phrases.add(phrase)

    return list(phrases)
```

```python
skill_anchor = "professional skill, technical ability, job competency"
```

```python
anchor_emb = model.encode(skill_anchor)
```

```python
def extract_relevant_skills(resume_text, job_description, threshold=0.34):
    phrases = extract_phrases(resume_text)

    if not phrases:
        return []

    phrase_embeddings = model.encode(phrases)
    job_embedding = model.encode(job_description)

    matched_skills = []

    for phrase, emb in zip(phrases, phrase_embeddings):
        score = cosine_similarity([emb], [job_embedding])[0][0]

        if score > threshold:
            matched_skills.append((phrase, score))

    matched_skills = sorted(
        matched_skills,
        key=lambda x: x[1],
        reverse=True
    )

    return matched_skills[:10]
```

```python
def filter_skills(phrases):
    skill_phrases = []

    phrase_embs = model.encode(phrases)

    for phrase, emb in zip(phrases, phrase_embs):
        score = cosine_similarity([emb], [anchor_emb])[0][0]

        if score > 0.4:  # threshold you can tune
            skill_phrases.append((phrase, float(score)))

    return sorted(skill_phrases, key=lambda x: x[1], reverse=True)
```

```python
def extract_dynamic_skills(resume_text):
    phrases = extract_phrases(resume_text)
    skills = filter_skills(phrases)
    return skills
```

```python
def deduplicate(skills, threshold=0.85):
    final = []

    for skill, score in skills:
        keep = True

        for fskill, _ in final:
            sim = cosine_similarity(
                model.encode([skill]),
                model.encode([fskill])
            )[0][0]

            if sim > threshold:
                keep = False
                break

        if keep:
            final.append((skill, score))

    return final
```
